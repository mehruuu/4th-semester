```python
train['CryoSleep'] = train['CryoSleep'].fillna(False)
test['CryoSleep'] = test['CryoSleep'].fillna(False)

train['HomePlanet'] = train['HomePlanet'].fillna('Unknown')
test['HomePlanet'] = test['HomePlanet'].fillna('Unknown')

train['Destination'] = train['Destination'].fillna('Unknown')
test['Destination'] = test['Destination'].fillna('Unknown')

train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age'] = test['Age'].fillna(test['Age'].median())
```

```python
train.drop(['Name', 'Cabin'], axis=1, inplace=True)
train.drop(['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck',"VIP"], axis=1, inplace=True)
```

```python
test.drop(['Name', 'Cabin'], axis=1, inplace=True)
test.drop(['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck',"VIP"], axis=1, inplace=True)
```

```python
train
```

```python
categorical_columns = ['HomePlanet', 'CryoSleep', 'Destination']
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
    label_encoders[col] = le
```

```python
X_train = train.drop('Transported', axis=1)
y_train = train['Transported'].astype(int)  
X_test = test
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)



```

```python
submission = pd.DataFrame({'PassengerId': X_test['PassengerId'], 'Transported': y_pred})
submission['Transported'] = submission['Transported'].astype(bool)  
submission.to_csv('submission.csv', index=False)
```
